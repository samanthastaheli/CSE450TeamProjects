## Charts

### Stats Data

```python
alt.Chart(label1_stats).mark_bar(opacity=0.85).encode(
    x="feature",
    y="recall"
).properties(
    width=300,
    height=200,
    title={'text': 'Economic Features Individual Recall'}
)
```

```python
alt.Chart(label1_stats).mark_bar(opacity=0.85).encode(
    x="feature",
    y="precision"
).properties(
    width=300,
    height=200,
    title={'text': 'Economic Features Individual Precision'}
)
```

```python
alt.Chart(label1_stats).mark_circle().encode(
    x="feature",
    y="precision"
).properties(
    width=300,
    height=200,
    title={'text': 'Economic Features Individual Precision'}
)
```

### Feature Imp Data

```python
alt.Chart(feature_imp__data).mark_bar(opacity=0.85).encode(
    x="features",
    y="empVarRatePos"
).properties(
    # width=300,
    # height=200,
    title={'text': 'emp.var.rate (positive) feature importance'}
)
```

```python
alt.Chart(feature_imp__data).mark_bar(opacity=0.85).encode(
    x="features",
    y="empVarRateNeg"
).properties(
    # width=300,
    # height=200,
    title={'text': 'emp.var.rate (negative) feature importance'}
)
```

```python
empVarRatePosChart = alt.Chart(feature_imp__data).mark_circle(color="#e76f51", opacity=0.8).encode(
    x="features",
    y="empVarRatePos",
).properties(
    # width=300,
    # height=200,
    title={'text': 'emp.var.rate feature importance'}
)

empVarRateNegChart = alt.Chart(feature_imp__data).mark_square(color="#1d3557", opacity=0.8).encode(
    x="features",
    y="empVarRateNeg",
).properties(
    # width=300,
    # height=200,
    title={'text': 'emp.var.rate feature importance', 'subtitle': 'orange circles: positive, blue squares: negative'}
)
```
